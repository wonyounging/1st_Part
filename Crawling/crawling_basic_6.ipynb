```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# https://www.siksinhot.com/search?keywords=%EC%8B%A0%EB%85%BC%ED%98%84%EC%97%AD
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
url = 'https://www.siksinhot.com/search?keywords=%EC%8B%A0%EB%85%BC%ED%98%84%EC%97%AD'
```

```python
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, )
```

```python
foodList = soup.find('div', attrs={'class':'div_search_menu'}).find_all('li')
```

```python
store = foodInfo.find('h2').get_text()
score = foodInfo.find('span').get_text()
place = foodInfo.find_all('a')[2].get_text()
menu = foodInfo.find_all('a')[3].get_text()
```

```python
print(store, score, place, menu)
```

```python
flist = []
for i in range(1):
    flist.append({
        '상호' : foodInfo.find('h2').get_text(),
        '평점' : foodInfo.find('span').get_text(),
        '위치' : foodInfo.find_all('a')[2].get_text(),
        '메뉴' : foodInfo.find_all('a')[3].get_text()
    })
flist
```

```python
for foodInfo in foodList:
    flist.append({
        '상호' : foodInfo.find('h2').get_text(),
        '평점' : foodInfo.find('span').get_text(),
        '위치' : foodInfo.find_all('a')[2].get_text(),
        '메뉴' : foodInfo.find_all('a')[3].get_text()
    })
flist
```

```python
df = pd.DataFrame(flist)
df.to_csv('맛집 크롤링.csv', index=False, encoding="CP949")
```

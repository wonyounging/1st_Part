```python
# 대학교 리스트를 데이터 프레임으로 전환
df = pd.read_excel('./data1/서울지역 대학교 위치.xlsx', engine = 'openpyxl')
df.head()
```

```python
# 서울지도 만들기
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start = 12,
                      tiles = 'Stamen Terrain')
```

```python
# 대학교 위치 정보를 Marker로 표시
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
```

```python
seoul_map
```

```python
seoul_map2 = folium.Map(location=[37.55, 126.98], zoom_start = 12,
                      tiles = 'Stamen Toner')
```

```python
# 대학교 위치 정보를 Marker로 표시
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat, lng], radius = 10,
                        color = 'brown', fill=True, fill_color = 'coral',
                        fill_opacity=0.7).add_to(seoul_map2)
```

```python
seoul_map2
```

```python
# 경기도 인구변화 데이터를 불러와서 데이터 프레임으로 변환
df2 = pd.read_excel('./data1/경기도인구데이터.xlsx', index_col='구분', engine='openpyxl')
df2.columns = df2.columns.map(str)
```

```python
df2.head(3)
```

```python
# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
import json

geo_path = './data1/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
```

```python
geo_data
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], tiles='Stamen Terrain',
                  zoomstart=9)
```

```python
g_map
```

```python
# 출력할 연도를 선택(2007~2017)
year = '2017'

# 단계 구분도(Choropleth) 단계 구분도 표시하기
folium.Choropleth(geo_data=geo_data, data=df2[year], columns=[df2.index, df2[year]],
                 fill_color='YlOrRd', fill_opacity=0.6, line_opacity=0.3,
                  threshold_scale=[10000,100000,300000,500000, 7000000],
                  key_on='feature.properties.name').add_to(g_map)
```

```python
g_map
```

```python
# 지도를 Html로 저장
g_map.save('./gyonggi_population' + year + '.html')
```

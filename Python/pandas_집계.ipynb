```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
# age를 평균으로 대체
titanic.age.fillna(titanic.age.mean(), inplace=True)
```

```python
# embarked 을 최빈값으로 대체
titanic.embarked.value_counts()
titanic.embarked.fillna('S', inplace=True)
```

```python
# embark_town 결측값이 있는 행 삭제
titanic.dropna(subset=['embark_town'], inplace=True)
```

```python
# deck 삭제
titanic.drop(['deck'], axis=1)
```

```python
# titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adult 이하면 child   (ㅣambda를 이용)
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult' if x> 20 else 'child')
```

```python
# bins = [1,20,30,50,70,100]
# label = '[미성년자', 청년, 중년, 장년, 노년]

bins = [1,20,30,50,70,100]
labels = ['미성년자', '청년', '중년', '장년', '노년']

# age를 기준으로 ;age_cut' 컬럼을 생성 후 두 컬럼을 출력하기
titanic['age_cat'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
def get_grade(x):
    if x > 90:
        grade = 'A'
    elif x >= 80:
        grade = 'B'
    elif x >= 70:
        grade = 'C'
    elif x >= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade
```

```python
df['grade'] = df['score'].apply(get_grade)
df = df.sort_values('name')
df
```

```python
# 데이터 프레임.pivot(index = 인덱스로 사용할 컬럼,
# columns = 컬럼으로 사용할 컬럼, values=값으로 사용할 컬럼)
# name, subject, score
df.pivot(index = 'name', columns = 'subject', values='score')
```

```python
df.pivot(index = 'name', columns = 'subject', values='grade')
```

# groupby

```python
# df.groupby(그룹 기준 컬럼).통계적용 컬럼.통계함수
# .count() : 누락 값을 제외한 데이터 수
# .size() : 누락 값을 포함한 데이터 수
# .mean(), .sum(), .std(), .min(), .max()
```

```python
# 객실 등급(pclass) 별 승선자 수 결과 집계
df_titanic1 = df_titanic.groupby('pclass').survived.count().to_frame()
df_titanic1
```

```python
# 객실 등급(pclass) 별 생존자 수 결과 집계
df_titanic2 = df_titanic.groupby('pclass').survived.sum().to_frame()
df_titanic2
```

```python
# 객실 등급(pclass) 별 생존율 결과 집계
df_titanic3 = df_titanic.groupby('pclass').survived.mean().to_frame().round(3)*100
df_titanic3
```

```python
# 객실 등급(pclass) 별 생존자 수 결과 집계
df_titanic4 = pd.concat([df_titanic1, df_titanic2, df_titanic3], axis = 1)
df_titanic4.columns = ['승전자 수', '생존자 수', '생존율(%)']
df_titanic4
```

```python
# 성별(sex) 별 승선자 수 결과 집계
df_titanic1 = df_titanic.groupby('sex').survived.count().to_frame()
df_titanic1
```

```python
# 성별(sex) 별 생존자 수 결과 집계
df_titanic2 = df_titanic.groupby('sex').survived.sum().to_frame()
df_titanic2
```

```python
# 성별(sex) 별 생존율 결과 집계
df_titanic3 = df_titanic.groupby('sex').survived.mean().to_frame().round(3)*100
df_titanic3
```

```python
# 성별(sex) 별 생존자 수 결과 집계
df_titanic4 = pd.concat([df_titanic1, df_titanic2, df_titanic3], axis = 1)
df_titanic4.columns = ['승전자 수', '생존자 수', '생존율(%)']
df_titanic4
```

```python
# 누락값(NaN)을 앞 데이터로 채움
df = df.fillna(method = 'ffill')  # ffill은 forward로 fill을 함
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]  # 한 줄 한 줄 확인하면서 하는 것이 좋음
df_seoul
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul
```

```python
df_seoul.set_index('전입지', inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
# 스타일 서식 지정
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
ax.plot(col_years, df3.loc['충청남도', :], marker='o', markerfacecolor='g', markersize=10, color='y', linewidth=2, label='서울->충남')
ax.plot(col_years, df3.loc['경상북도', :], marker='o', markerfacecolor='b', markersize=10, color='skyblue', linewidth=2, label='서울->경북')
ax.plot(col_years, df3.loc['강원도', :], marker='o', markerfacecolor='m', markersize=10, color='r', linewidth=2, label='서울->강원')

ax.legend(loc=0)
ax.set_title('서울->충남, 경북, 강원 인구이동', size=20)
ax.set_xlabel('기간', size=12)
ax.set_ylabel('인구이동수', size=12)
ax.set_xticklabels(col_years, rotation=90)
plt.show()
```

# 남북한 전력량으로 secondary_y를 그리기

```python
df = pd.read_excel('./data1/남북한발전전력량.xlsx', engine='openpyxl', convert_float=True)
# 북한 데이터만 가져오기
df = df.loc[5:9]
df.head(2)
```

```python
df.drop('전력량 (억㎾h)', axis = 'columns', inplace=True)
df.head(2)
```

```python
# 인덱스를 '발전 전력별'로 변경
df.set_index('발전 전력별', inplace=True)
df.head(2)
```

```python
df = df.T
df.head(2)
```

```python
# 증감율 (변동율) 계산
df = df.rename(columns={'합계':'총발전량'})
df.head(2)
```

```python
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df.head(3)
```

```python
df['증감율'] = ((df['총발전량']/df['총발전량 - 1년']) - 1) * 100
df.head(3)
```

```python
# 2축 그래프 그리기
ax1 = df[['수력', '화력']].plot(kind='bar', figsize=(20,10),width=0.7, stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감율, ls='--', marker='o', markersize=10,   # ls는 라인 스타일(line style)
        color='g', label='전년대비 증감율(%)')
ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size=30)
ax1.set_xlabel('발전량(억 Kwh)')
ax2.set_xlabel('전년대비 증감율')

plt.title('북한 전력 발전량(1990-2016)', size=25)
ax1.legend(loc='upper left', fontsize = 20)
plt.show()
```

```python
# 히스토그램 - 단변수 데이터의 빈도수를 나타냄
df = pd.read_csv('./data1/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model_year', 'origin', 'name']
```

```python
plt.style.use('ggplot')
# 연비(mpg) 열에 대한 히스토그램을 그리기
df['mpg'].plot(kind = 'hist', bins = 10) # bins = 10dms mpg10-50 정의 값을 10구간으로 나눈다.
#그래프 꾸미기
plt.title('histogram')
plt.xlabel('mpg')
plt.show()
```

```python
# 산정도 scatter
plt.style.use('Solarize_Light2')
# plt.style.available : 스타일 확인 코드
df.plot(x = 'weight', y = 'mpg', kind = 'scatter', c = 'coral', s = 10, figsize=(10,5))
plt.title('scatter plot mpg VS. weight')
plt.show()
```

```python
plt.style.available
```

```python
# 버블 차트
plt.style.use('bmh')
# 버블차트를 그리기 위해 cylinder 개수의 상대적 비율을 계산하여 시리슺 생성
cylinder_size = df.cylinders / df.cylinders.max() * 300
df.plot(x = 'weight', y = 'mpg', kind = 'scatter', c = 'coral', s = cylinder_size, figsize=(10,5),
       alpha=0.3)
plt.title('scatter plot : mpg-weight')
plt.show()
```

```python
plt.style.use('default')
# 3개의 변수로 산점도 그리기

df.plot(x = 'weight', y = 'mpg', kind = 'scatter', c = cylinder_size, s=50, figsize=(10,5),
        marker='+', alpha=0.3, cmap='jet') # cmap은 colo map 'viridis','jet', 'gray', 'hot'
plt.title('scatter plot : mpg-weight')
plt.show()
```

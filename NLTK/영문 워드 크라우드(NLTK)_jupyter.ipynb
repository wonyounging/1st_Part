# 영문 워드 크라우드(NLTK)

```python
# NLTK(Naturl Language ToolKit)
import nltk
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image # 그림을 불러오는 패키지
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

# 1. stopwords(불용어)처리

```python
sw_nltk = stopwords.words('english')
print(sw_nltk)
print(len(sw_nltk))
```

```python
# wordCloud에서 제공하는 불용어
sw_wc = STOPWORDS
print(sw_wc)
print(len(sw_wc))
```

# 2. 이상한 나라 엘리스 분석

```python
with open('./data/Alice/Alice.txt') as file:
    text = file.read()
```

```python
text
```

```python
# 'said' 단어를 불용어 처리
sw_wc.add('said')
len(sw_wc)
```

```python
wc = WordCloud(background_color="white", max_words=2000, stopwords=sw_wc)
wc = wc.generate(text)
```

```python
print(wc)
```

```python
count = 0
for key, value in wc.words_.items():
    print(key, ':', value)
    count += 1
    if count > 10:
        break
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt. show()
```

```python
mask = np.array(Image.open('./data/Alice/Alice_mask.png'))
```

# Star Wars

```python
# 데이터 불러오기
with open('./data/Alice/A_new_hope.txt') as file:
    text = file.read()
```

```python
text = text.replace('HAN', 'Han').replace("LUKE'S", "Luke")
```

```python
# 불용어 처리
sw_wc = STOPWORDS
sw_wc.add('int')
sw_wc.add('ext')
len(sw_wc)
```

```python
wc = WordCloud(background_color="white", max_words=1000, stopwords=sw_wc)
wc = wc.generate(text)
```

```python
count = 0
for key, value in wc.words_.items():
    print(key, ':', value)
    count += 1
    if count > 10:
        break
```

```python
mask = np.array(Image.open('./data/Alice/Stormtrooper_mask.png'))
```

```python
wc = WordCloud(background_color="white", max_words=1000, stopwords=sw_wc, mask=mask).generate(text)
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt. show()
```

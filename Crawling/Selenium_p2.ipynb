```python
# 순위 카테고리 채널 구독자수 조회수 비디오수
# 셀레늄으로 해당페이지를 열고 + BS4를 사용해서 크롤링
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import pandas as pd
```

# 셀레늄 이용 페이지 열기

```python
url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)
```

# BS4 이용 크롤링

```python
# soup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
trs = soup.select('tr.aos-init')
len(trs)
```

```python
tr = trs[0]
tr
```

```python
# 순위
int(tr.select_one('.rank').get_text().strip())
```

```python
# 카테고리
tr.select_one('.category').get_text().strip()[1:-1]
```

```python
# 채널명
# tr.select('a')[1].get_text().strip().replace("\n", "").replace("\t", "")
tr.select_one('.subject a').get_text().strip()
```

```python
# 구독자수
tr.select_one('.subscriber_cnt').get_text().strip()
```

```python
# 조회수
tr.select_one('.view_cnt').get_text().strip()
```

```python
# 비디오수
tr.select_one('.video_cnt').get_text().strip()
```

```python
# 첫 페이지 크롤링
lines = []
for tr in trs:
    rank = int(tr.select_one('.rank').get_text().strip())
    category = tr.select_one('.category').get_text().strip()[1:-1]
    name = tr.select_one('.subject a').get_text().strip()
    subscriber = tr.select_one('.subscriber_cnt').get_text().strip()
    view = tr.select_one('.view_cnt').get_text().strip()
    video = tr.select_one('.video_cnt').get_text().strip()
    
    lines.append([rank, category, name, subscriber, view, video])
lines
```

```python
df = pd.DataFrame(lines, columns=['순위', '카테고리', '채널명', '구독자수', '조회수', '비디오수'])
df
```

```python
# 숫자로 바꿔주는 함수
def convert_unit(s):
    s = s.replace("억", "").replace("개", "").replace(",", "").replace("만", "0000")
    return f'{int(s):,d}'
```

```python
convert_unit('303억4016만')
```

```python
for column in df.columns[3:]:
    df[column] = df[column].apply(convert_unit)
df.head()
```

```python
from tqdm.notebook import tqdm
```

```python
lines = []
for page in tqdm(range(1,11)):
    url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=' + str(page)
    driver.get(url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    trs = soup.select('tr.aos-init')
    
    for tr in trs:
        rank = int(tr.select_one('.rank').get_text().strip())
        category = tr.select_one('.category').get_text().strip()[1:-1]
        name = tr.select_one('.subject a').get_text().strip()
        subscriber = tr.select_one('.subscriber_cnt').get_text().strip()
        view = tr.select_one('.view_cnt').get_text().strip()
        video = tr.select_one('.video_cnt').get_text().strip()

        lines.append([rank, category, name, subscriber, view, video])
```

```python
df = pd.DataFrame(lines, columns=['순위', '카테고리', '채널명', '구독자수', '조회수', '비디오수'])
for column in df.columns[3:]:
    df[column] = df[column].apply(convert_unit)
```
